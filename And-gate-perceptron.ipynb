```python
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
Y = np.array([0, 0, 0, 1])
```

```python
weights = np.zeros(2)
bias = 0
learning_rate = 0.2
epochs = 8
```

```python
def step_function(x):
    return 1 if x >= 0 else 0
```

```python
for epoch in range(epochs):
    for i in range(len(X)):
        print(X[i])
        print(weights)
        print(bias)
        linear_output = np.dot(X[i], weights) + bias
        print("Linear_output",linear_output)
        prediction = step_function(linear_output)
        print("Prodiction",prediction)

        error = Y[i] - prediction


        weights = weights + learning_rate * error * X[i]
        print("New W",weights)
        bias = bias + learning_rate * error
        print("New bias",bias)
        print("\n")
```

```python
def predict(x):
    return step_function(np.dot(x, weights) + bias)

print("Testing:")
for x in X:
    print(f"Input: {x} → Output: {predict(x)}")
```
